## Which universities do they enroll? What are the locations?

Create a map using data from [the Ministry of Education and Science of Ukraine].(https://web.archive.org/web/20220211073341/https://studyinukraine.gov.ua/en/life-in-ukraine/international-students-in-ukraine/)




```python
import geocoder
```

```python
uni= pd.read_csv("ukraine_top10_uni.csv")
uni
```

```python
uni_list = uni.name.to_list()

uni_location =[]

for loc in uni_list:
    location = loc
    ret = geocoder.osm(location)
    latlng = ret.latlng
    uni_location.append(latlng)
uni_location
```

```python
uni["latlng"] = uni_location
uni
```

```python
uni.to_csv("ukraine_top10_uni.csv", index = False)
```
